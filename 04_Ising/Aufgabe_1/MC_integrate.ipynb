```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def rng(mean, stdev, N):
    return np.random.normal(mean, stdev, N)

def integrand(x):
    return np.exp(0.5*-x**2)
```

```python
## integrate

sampleses = [1000, 100000, 10000000, 100000000]
mean = 0
stdev = 1

for samples in sampleses:
    plt.figure()
    data = rng(mean, stdev, samples)

    plt.hist(data, bins=1000, density=True)
    plt.show()

    integral = np.sum(integrand(data))/samples * 2*np.pi**0.5
    print(f'Integral result for {samples} samples: {integral:.6f}')
```
