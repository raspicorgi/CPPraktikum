```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

trial_counts = [500, 10000, 50000, 100000]
fig, axs = plt.subplots(1, len(trial_counts), figsize=(16, 5))
fig.tight_layout()

for ax, trials in zip(axs, trial_counts):
    points = np.random.rand(trials, 2)
    points_in_circle = points[np.linalg.norm(points, axis=1) < 1]
    points_outside_circle = points[np.linalg.norm(points, axis=1) >= 1]
    ax.plot(points_outside_circle[:, 0], points_outside_circle[:, 1], 'ro')
    ax.plot(points_in_circle[:, 0], points_in_circle[:, 1], 'bo')
    # ax.set_xlim(0, 1)
    # ax.set_ylim(0, 1)
    ax.set_aspect('equal', adjustable='box')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title(f'{trials} Stichproben\nErmittelter Wert fr Pi: {4 * len(points_in_circle) / trials:.5f}')
    ax.legend(['Auerhalb', 'Innerhalb'], loc='lower left')

plt.suptitle(r'Monte Carlo Bestimmung von Pi mit verschiedenen Versuchen')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])


# print("Estimate of pi:", 4 * len(points_in_circle) / num_trials)
```

Wir betrachten das Einheitsquadrat. Ein Punkt liegt dann im Einheitskreis, wenn  
$x^2 + y^2 < 1$.  
Wir erhalten $\pi$ wie folgt:  
$A_\text{Kreis} = \pi r^2 = \pi$ und somit $\frac{A_\text{Viertelkreis}}{A_\text{Quadrat}} = \frac{\pi}{4}$.
Fr das Verhltnis $A_\text{Viertelkreis}/A_\text{Quadrat}$ wird das Verhltnis der Anzahl der Punkte verwendet.
