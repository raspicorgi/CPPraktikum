```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def rng(mean, stdev, N):
    """
    Generate an array of random numbers following a normal distribution.

    Parameters:
    mean (float): The mean of the normal distribution.
    stdev (float): The standard deviation of the normal distribution.
    N (int): The number of random numbers to generate.

    Returns:
    numpy.ndarray: An array of N random numbers following the specified normal distribution.
    """
    return np.random.normal(mean, stdev, N)

def integrand(x):
    return np.exp(0.5*-x**2)
```

```python
## integrate

samples = 10000000
mean = 0
stdev = 1

data = rng(mean, stdev, samples)

plt.hist(data, bins=1000, density=True)
plt.show()

integral = np.sum(integrand(data))/samples * 2*np.pi**0.5
print(f'Integral result: {integral}')
```
